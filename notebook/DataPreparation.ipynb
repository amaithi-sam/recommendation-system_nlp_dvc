## Text preprocessing

```python
import re 
```

```python
df2 = df.copy()

rmv_spc = lambda a:a.strip()

get_list = lambda a:list(map(rmv_spc, re.split('& |, |\*', a)))
```

```python
list(re.split('& |, |\*', 'Kitchen, Garden & Pets'))
```

```python
list(map(rmv_spc, re.split('& |, |\*', 'Kitchen, Garden & Pets')))
```

```python

for col in ['category', 'sub_category', 'type']:
    df2[col] = df2[col].apply(get_list)
```

```python

def cleaner(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''

for col in ['category', 'sub_category', 'type', 'brand']:
    df2[col] = df2[col].apply(cleaner)

def couple(x):
    return ' '.join(x['category']) + ' ' + ' '.join(x['sub_category']) + ' '+x['brand']+' ' +' '.join( x['type'])

df2['product_classification_features'] = df2.apply(couple, axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel, cosine_similarity
```

```python
count_vector = TfidfVectorizer(stop_words='english')

count_matrix = count_vector.fit_transform(df3['product_classification_features'])
```

```python
def get_recommendations_2(title, cosine_sim=cosine_sim):
    
    idx = indices[title][0]

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:11]

    movie_indices = [i[0] for i in sim_scores]

    return df4[['product', 'rating', 'sale_price']].iloc[movie_indices]
```
